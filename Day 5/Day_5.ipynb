```python
'''Creating genai_utils.py with utility functions for GenAI applications:'''

%%writefile genai_utils.py
import json
from typing import List, Dict, Any

def chunk_text(text: str, chunk_size: int = 20, overlap: int = 5) -> List[str]:
    """Splits long text into overlapping chunks for Vector DB indexing."""
    words = text.split()
    chunks = []
    for i in range(0, len(words), chunk_size - overlap):
        chunk = " ".join(words[i:i + chunk_size])
        if chunk:
            chunks.append(chunk)
    return chunks

def format_agent_prompt(system_msg: str, user_query: str, context: str = "") -> str:
    """Injects dynamic context into a strict prompt template for agents."""
    prompt = f"System: {system_msg}\n"
    if context:
        prompt += f"Context: {context}\n"
    prompt += f"User: {user_query}\nAgent:"
    return prompt

def write_to_jsonl(data: List[Dict[str, Any]], filename: str) -> None:
    """Exports dictionaries to a JSONL file for batch processing/fine-tuning."""
    with open(filename, 'w', encoding='utf-8') as f:
        for item in data:
            f.write(json.dumps(item) + '\n')
```

```python
'''Import & Define Inputs
Load our new module and set up sample prompts for an analytical agent.'''

import genai_utils

print("✅ genai_utils imported successfully.")

# Sample inputs for an agentic analysis bot
system_role = "You are an analytical agent."
sample_data = [
    {
        "query": "Summarize the Q3 performance.", 
        "context": "Revenue up 15%, costs down 5%. New product launch successful."
    },
    {
        "query": "Analyze moving averages for this asset.", 
        "context": "The 50-day SMA has just crossed above the 200-day SMA."
    }
]
```

```python
'''Processing & Generating JSONL
Transform the inputs into a fine-tuning format and save as JSONL.'''

fine_tune_dataset = []

for item in sample_data:
    # Test the chunker (simulating long document processing)
    chunks = genai_utils.chunk_text(item["context"], chunk_size=5, overlap=1)
    
    # Format for standard OpenAI JSONL fine-tuning structure
    record = {
        "messages": [
            {"role": "system", "content": system_role},
            {"role": "user", "content": f"Context: {item['context']} Query: {item['query']}"},
            {"role": "assistant", "content": "Simulated ideal target response here."}
        ]
    }
    fine_tune_dataset.append(record)

# Export to JSONL
output_file = "agent_training_data.jsonl"
genai_utils.write_to_jsonl(fine_tune_dataset, output_file)

print(f"✅ Generated {len(fine_tune_dataset)} records in '{output_file}'")
```
